```python
import PyPDF2
```

```python
path_pdf = "/home/etienneguevel/Documents/SCAI/AI_TRIOMPH/documents/MESURE DES ACTIVITES CANCER SOUMISES A SEUILS_2023.pdf"
f = open(path_pdf, 'rb')
pdfReader = PyPDF2.PdfReader(f)

text = [pdfReader.pages[i].extract_text() for i in range(30,45)]
f.close()
```

```python
print(text[-1])
```

```python
import re

regex_cim = r'[A-Z][0-9\+]{2,5}'
regex_test = rf'\b{regex_cim}[^\n]*'
cim_codes = []

for p in text:
    cim_codes.extend(re.findall(regex_test, p))
```

```python
dict_cim_cancer = {re.search(regex_cim, c)[0]: re.split(regex_cim, c)[-1] for c in cim_codes}
```

```python
import pickle

with open("/home/etienneguevel/Documents/SCAI/AI_TRIOMPH/ai_triomph/triomph/src/cancer_cim.pkl", "wb") as f:
    pickle.dump(dict_cim_cancer, f)
```
